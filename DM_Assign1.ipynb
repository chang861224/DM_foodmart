```python
import json

### Load Transactions
print('Transactions Loading....')

f = open('transaction.json', 'r')

lines = f.readlines()
transaction = []

for line in lines:
    transaction.append(json.loads(line)['products'])

f.close()

print('Transactions Loaded!!')
```

```python
from mlxtend.preprocessing import TransactionEncoder

### Transaction Encoder
print('Transactions Encoding....')

te = TransactionEncoder()
te_array = te.fit(transaction).transform(transaction)
print(te_array)
print('{} x {}'.format(str(len(te_array)), str(len(te_array[0]))))

print('Transactions Encoded!!')
```

```python
import pandas as pd

df = pd.DataFrame(te_array, columns=te.columns_)
```

```python
df
```

```python
from mlxtend.frequent_patterns import fpgrowth, apriori
import numpy as np

frequent_itemsets = fpgrowth(df, min_support=0.0001, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

ar = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.9)
ar
```
